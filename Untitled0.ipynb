```python
def define_status(age):
  if age < 23:
    return '–º–æ–ª–æ–¥–æ–π'
  elif age < 60:
    return '–≤–∑—Ä–æ—Å–ª—ã–π'
  else:
    return '–ø–µ–Ω—Å–∏–æ–Ω–µ—Ä'

age = input('–í–≤–µ–¥–∏—Ç–µ —Å–≤–æ–π –≤–æ–∑—Ä–∞—Å—Ç:')
print(type(age))
print(f'–í–∞—à –≤–æ–∑—Ä–∞—Å—Ç: {age}, –≤—ã {define_status(int(age))}')
```

```python
import random

#–Ω–∞—Å—Ç—Ä–æ–π–∫–∞ –¥–∏–∞–ø–∞–∑–æ–Ω–∞ –∏ –º–∞–∫—Å–∏–º–∞–ª—å–Ω–æ–≥–æ –∫–æ–ª-—Ç–≤–∞ –ø–æ–ø—ã—Ç–æ–∫
max_attempts = 7
min_range = 1
max_range = 100

secret_number = random.randint(min_range, max_range)
attempts = 1

print(f"–Ø –∑–∞–≥–∞–¥–∞–ª —á–∏—Å–ª–æ –æ—Ç {min_range} –¥–æ {max_range}")
print(f"–£ –≤–∞—Å –µ—Å—Ç—å {max_attempts} –ø–æ–ø—ã—Ç–æ–∫!")

while True:

  guess = int(input("–í–∞—à–µ –ø—Ä–µ–¥–ø–æ–ª–æ–∂–µ–Ω–∏–µ: "))

  if guess == secret_number:
     print(f"\nüéâ –ü–æ–∑–¥—Ä–∞–≤–ª—è—é! –í—ã —É–≥–∞–¥–∞–ª–∏ —á–∏—Å–ª–æ {secret_number}!")
     print(f"üìä –ò—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–æ –ø–æ–ø—ã—Ç–æ–∫: {attempts}")
     break
  elif guess < secret_number:
      print("üìà –ó–∞–≥–∞–¥–∞–Ω–Ω–æ–µ —á–∏—Å–ª–æ –ë–û–õ–¨–®–ï")
  else:
      print("üìâ –ó–∞–≥–∞–¥–∞–Ω–Ω–æ–µ —á–∏—Å–ª–æ –ú–ï–ù–¨–®–ï")


  if attempts < max_attempts:
    print(f"üí° –û—Å—Ç–∞–ª–æ—Å—å –ø–æ–ø—ã—Ç–æ–∫: {max_attempts - attempts}")
    attempts +=1
  else:
    print(f"\nüíÄ Game Over! –í—ã –∏—Å—á–µ—Ä–ø–∞–ª–∏ –≤—Å–µ –ø–æ–ø—ã—Ç–∫–∏.")
    print(f"üî¢ –ó–∞–≥–∞–¥–∞–Ω–Ω–æ–µ —á–∏—Å–ª–æ –±—ã–ª–æ: {secret_number}")
    break
```
